1.  What is NumPy, and why is it widely used in Python

2. How does broadcasting work in NumPy
Ans 2. Broadcasting in NumPy is a powerful feature that allows arrays with different shapes to be used together in arithmetic operations without requiring explicit replication of data. When performing operations on arrays of unequal dimensions, NumPy automatically "broadcasts" the smaller array across the larger one so that their shapes become compatible. This is done by virtually stretching the dimensions of the smaller array without actually copying data, which makes operations memory-efficient and faster. Broadcasting follows a set of rules: starting from the trailing dimensions, NumPy compares shapes and allows operations if the dimensions are equal or if one of them is 1. If the shapes cannot be made compatible, a `ValueError` is raised. This feature simplifies coding and enables efficient vectorized computations, making NumPy highly effective for numerical and scientific computing.

4. Explain the use of the groupby() method in Pandas
Ans 4. The `groupby()` method in Pandas is a powerful tool used to group and aggregate data based on one or more keys (columns). Internally, it follows a "split-apply-combine" strategy: it splits the data based on the grouping key(s), applies a function to each group, and combines the results into a new DataFrame or Series. Overall, `groupby()` is essential for data summarization, transformation, and analysis in Pandas.

6. What are the differences between NumPy arrays and Python lists
Ans 6. NumPy arrays and Python lists are both used to store collections of data, but they differ significantly in functionality, performance, and memory efficiency. A primary difference is that NumPy arrays are designed for numerical and scientific computing, offering faster execution and lower memory consumption than Python lists, especially for large datasets. Unlike Python lists, which can hold elements of mixed data types, NumPy arrays are homogeneous, meaning all elements must be of the same type, enabling optimized storage and operations. NumPy also supports vectorized operations, allowing element-wise computations without the need for explicit loops, which significantly boosts performance. Additionally, NumPy provides a rich set of functions for mathematical, statistical, and linear algebra operations, which are either not available or require more effort to implement with standard Python lists. In summary, while Python lists are more flexible for general-purpose programming, NumPy arrays are the preferred choice for high-performance numerical tasks.

8.What does the term “vectorized operation” mean in NumPy
Ans 8. In NumPy, the term “vectorized operation” refers to the process of performing operations on entire arrays (or vectors) of data without the explicit use of loops. Instead of iterating through individual elements, NumPy applies operations at the array level using highly optimized C code under the hood. This approach significantly enhances performance, readability, and efficiency when handling large datasets. For example, adding two NumPy arrays with `a + b` automatically performs element-wise addition across the entire arrays, unlike traditional Python where you'd need a loop. Vectorized operations take advantage of NumPy's internal optimizations and broadcasting rules, allowing for faster and more concise code execution, especially useful in data science, machine learning, and scientific computing tasks.

15.  How does NumPy handle multidimensional arrays
Ans 15. NumPy handles multidimensional arrays through its powerful `ndarray` (n-dimensional array) object, which allows efficient storage and manipulation of large datasets in multiple dimensions. Internally, multidimensional arrays are stored in contiguous memory blocks, enabling fast access and computation using vectorized operations, which are much faster than traditional Python loops. This design makes NumPy a cornerstone library for scientific computing and data analysis in Python.

17.  Explain the difference between apply() and map() in Pandas
Ans 17. In Pandas, both `apply()` and `map()` are used to apply functions to data, but they differ in scope and usage. The `map()` function is primarily used with **Series** objects and is best suited for element-wise operations, such as transforming or mapping values using a function, dictionary, or series. For example, it’s often used to replace or convert values in a single column. On the other hand, `apply()` is more versatile and can be used with both Series and DataFrame objects. When applied to a DataFrame, `apply()` can operate along an axis (rows or columns) and can handle row-wise or column-wise computations, making it suitable for more complex transformations. In summary, use `map()` for simple value-wise operations on a Series, and `apply()` for more flexible row/column-wise operations or custom logic across DataFrames.

21. Why is NumPy’s array slicing faster than Python’s list slicing
Ans 21. NumPy’s array slicing is significantly faster than Python’s built-in list slicing because NumPy arrays are implemented in C and use contiguous blocks of memory, enabling efficient indexing and access. Unlike Python lists, which are collections of references to separate memory locations (making slicing a more resource-intensive operation), NumPy arrays store homogeneous data types in a compact structure. This allows slicing operations to return views of the original data rather than copies, avoiding unnecessary memory allocation and data duplication. Additionally, NumPy leverages vectorized operations and low-level optimizations that are much closer to machine-level execution, resulting in faster computation and reduced overhead compared to Python's more flexible but slower object-oriented list handling.

```python
#4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame
data = {
    'Math': [90, 85, 70, 65, 95],
    'Science': [88, 82, 72, 60, 96],
    'English': [75, 78, 72, 70, 80],
    'History': [80, 77, 69, 65, 85]
}

df = pd.DataFrame(data)

# Calculate the correlation matrix
corr_matrix = df.corr()

# Set the plot style
sns.set(style="white")

# Create the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5, square=True)

# Add a title
plt.title("Correlation Matrix Heatmap")
plt.show()
```
